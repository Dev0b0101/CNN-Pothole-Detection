```python
import cv2
from matplotlib import pyplot as plt
import os
import random
# create figure
fig = plt.figure(figsize=(10, 7))
# setting values to rows and column variables
rows = 2
columns = 1
#Two list of image
pothole_path = '/Users/xiezhenyi/Documents/GitHub/mindspark-hackathon-Huixian-Gong/Dataset/Pothole'
normal_path = '/Users/xiezhenyi/Documents/GitHub/mindspark-hackathon-Huixian-Gong/Dataset/Normal'
image_norm = os.listdir(normal_path)
image_pot = os.listdir(pothole_path)
pot_example= random.choice(image_pot)
norm_example = random.choice(image_norm)
Image_1 = cv2.imread("/Users/xiezhenyi/Documents/GitHub/mindspark-hackathon-Huixian-Gong/Dataset/Pothole/"+pot_example)
Image_2 = cv2.imread("/Users/xiezhenyi/Documents/GitHub/mindspark-hackathon-Huixian-Gong/Dataset/Normal/"+norm_example)
fig.add_subplot(rows,columns,1)
plt.imshow(Image_1)
plt.axis('off')
plt.title("Pothole_example")
fig.add_subplot(rows,columns,2)
plt.imshow(Image_2)
plt.axis('off')
plt.title("Normal_example")
```
